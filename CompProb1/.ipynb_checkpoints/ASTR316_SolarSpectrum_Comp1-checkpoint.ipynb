```python
#Read in the solar spectrum, which is just a set of flux measurements at specific wavelengths
SolarWavelength, SolarFlux = np.loadtxt('SolarSpectrum.tbl', usecols=range(2), skiprows=3, unpack=True)
```

```python
#now make a figure that shows the solar spectrum.
#use log axes in both x and y, and set the x axis to go from 0.1 to 10 microns so that it matches Figure 2.1
plt.loglog(SolarWavelength,SolarFlux,linewidth = 2)
plt.xlim(0.1,10)
plt.ylim(10,1e7)
xlabel = plt.xlabel('Wavelength ($\mu$m)') 
ylabel = plt.ylabel('Flux density (ergs s${^-1}$ cm$^{-2}$ $\mu$m${^-1}$)');
```

The plot above matches the format of Figure 2.1 in the text, so now to make a plot that matches the format of 2.6, we need to adjust the x axis to show the region from 0.4 to 1 microns, on a linear scale.

```python
#now make a figure that shows the solar spectrum.
#use log axes in both x and y, and set the x axis to go from 0.1 to 10 microns so that it matches Figure 2.1
plt.semilogx(SolarWavelength,SolarFlux,linewidth = 2)
plt.xlim(0.4,1)
plt.ylim(0,2.5e6)
xlabel = plt.xlabel('Wavelength ($\mu$m)') 
ylabel = plt.ylabel('Flux density (ergs s${^-1}$ cm$^{-2}$ $\mu$m${^-1}$)');
```

```python
#use trapezoidal integration to integrate the solar spectrum
bolometric_flux = np.trapz(SolarFlux, SolarWavelength)
print(bolometric_flux)
```

```python
#calculate the solar luminosity as L = 4 pi d^2 bolometric_flux
pi = 3.1415
au = 1.5e13

Solar_L = 4 * pi * au**2 * bolometric_flux
print(Solar_L)
```

```python
#define a function that produces a blackbody spectrum
#The function will need to be given 2 inputs: an array of wavelengths (in meters!) 
#that it should compute the intensities of, and the temperature (in degrees Kelvin) of the blackbody.
#The function can then calculate the Planck function based on those inputs and the necessary physical 
#constants; the function can then output an array with intensities for each wavelength it was given.

#the function, with comments on each line to remind ourselves what the units of the constants are:
def planck(wavelengths, temperature):
    c = 2.998E8 #speed of light in m/s
    h = 6.626E-34 #Plank constant 
    k = 1.381E-23 #Boltzmann constant
    return 2*h*c**2 / (wavelengths**5 * (np.exp(h*c/ (wavelengths*k*temperature))-1) )
```

```python
#Now create an array of wavelengths that we can use the Planck Function to compute intensities for.
#Since the Planck function 
#once we've defined those arrays, print out a few elements to check that we've computed them correctly
wavelengths = np.linspace(1E-8,1E-3,3000)  #what does this command do? 
print(wavelengths[0], wavelengths[-1], wavelengths[-2])
```

```python
#now create an array of a few temperatures that we want to plot blackbody functions for
temperatures = np.array([100, 1800, 4000, 5775, 10000])
#print temperatures[0]
```

```python
# Now let's use equation 5.90 in Ryden and Peterson to calculate the intensity of blackbody radiation 
# per unit wavelength for each temperature: 
planck_lambda_100 = planck(wavelengths, temperatures[0])
planck_lambda_1800 = planck(wavelengths, temperatures[1])
planck_lambda_4000 = planck(wavelengths, temperatures[2])
planck_lambda_5700 = planck(wavelengths, temperatures[3])
planck_lambda_10000 = planck(wavelengths, temperatures[4])
```

```python
#now make a figure that shows the solar spectrum.
#use log axes in both x and y, and set the x axis to go from 0.1 to 10 microns so that it matches Figure 2.1
plt.loglog(SolarWavelength,SolarFlux,linewidth = 2)
plt.loglog(wavelengths[:]*1E6,planck_lambda_5700[:]*0.7e-7,linewidth = 1)
plt.xlim(0.1,10)
plt.ylim(10,1e7)
xlabel = plt.xlabel('Wavelength ($\mu$m)') 
ylabel = plt.ylabel('Flux density (ergs s$^{-1}$ cm$^{-2}$ $\mu$m${^-1}$)');
```

```python
#Now let's plot each of these blackbody spectra.

#We'll use the pyplot command from the matplotlib library 
#to make these plots; when we imported the pyplot package earlier, 
#we gave it a new, shorter name ('plt'): otherwise we'd have to call 
#these commands like matplotlib.pyplot.loglog or matplotlib.pyplot.axis, 
#which would get old quickly.plt.semilogy(wavelengths[:]*1E6,planck_lambda_100[:],linewidth = 2)
xlabel = plt.xlabel('Wavelength (microns)') 
ylabel = plt.ylabel('Log flux (W m^-2 micron^-1)')
plt.semilogy(wavelengths[:]*1E6,planck_lambda_1800[:],linewidth = 1)
plt.semilogy(wavelengths[:]*1E6,planck_lambda_4000[:],linewidth = 1)
plt.semilogy(wavelengths[:]*1E6,planck_lambda_5700[:],linewidth = 1)
plt.semilogy(wavelengths[:]*1E6,planck_lambda_10000[:],linewidth = 1)
plt.axis([1E-2,2,5E5,1E15])
legend = plt.legend(temperatures, loc=4)
p = plt.axvspan(0.4, 0.8, facecolor='0.8', alpha=0.5)
```

Hmmm.  Those functions look pretty jagged at short wavelengths, even though we are calculating the wavelength arrays with 3000 datapoints.  Why aren't 3000 datapoints enough to produce a smooth curve? 

Let's see if we can't do a better job by using another command to make our initial array of wavelengths...

```python
#recreate our wavelength and  arrays with a different array creation command
wavelengths = np.logspace(-8,-3,300)  #what does this command do? 
```

And then compute and plot the Planck functions with *exactly* the same code as we used in the two cells above...

```python
# recompute the blackbody curves using this new 'wavelengths' array 
planck_lambda_100 = planck(wavelengths, temperatures[0])
planck_lambda_1800 = planck(wavelengths, temperatures[1])
planck_lambda_4000 = planck(wavelengths, temperatures[2])
planck_lambda_5700 = planck(wavelengths, temperatures[3])
planck_lambda_10000 = planck(wavelengths, temperatures[4])

#and remake the figure too.
xlabel = plt.xlabel('Wavelength (microns)') 
ylabel = plt.ylabel('Log flux (W m^-2 micron^-1)')
plt.semilogy(wavelengths[:]*1E6,planck_lambda_1800[:],linewidth = 1)
plt.semilogy(wavelengths[:]*1E6,planck_lambda_4000[:],linewidth = 1)
plt.semilogy(wavelengths[:]*1E6,planck_lambda_5700[:],linewidth = 1)
plt.semilogy(wavelengths[:]*1E6,planck_lambda_10000[:],linewidth = 1)
plt.axis([1E-2,2,5E5,1E15])
legend = plt.legend(temperatures, loc=4)
p = plt.axvspan(0.4, 0.8, facecolor='0.8', alpha=0.5)
```

That looks a lot better!
