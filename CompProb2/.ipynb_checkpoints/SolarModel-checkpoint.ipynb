### 4. Calculate numerical estimates of dP/dr and -GM $\rho$(r) /r$^2$ as a function of position within the stellar interior

```python
n_profile = len(radius)
dP_dr = np.zeros(n_profile)

GMrho_over_rsquared = np.zeros(n_profile)

for x in range(0,n_profile-1):

    if x == 0: 
        dP_dr[x] = 0
        GMrho_over_rsquared[x] = 0
    else:
        dP_dr[x] = (10.**logP[x-1] - 10.**logP[x]) / (Rsun*(radius[x-1] - radius[x])) 
        GMrho_over_rsquared[x] = - bigG * ((mass[x-1] + mass[x])/2) * Msun * 10.**((log_rho[x] + log_rho[x-1])/2.) / ( ( ((radius[x-1] + radius[x])/2)*Rsun)**2.)
```

### 5. Create a figure that uses your values for  dP/dr and -GM $\rho$(r) /r$^2$  to show that the vast majority of the material within this star is in hydrostatic equilibrium 

```python
plt.plot(radius, dP_dr)
plt.plot(radius, GMrho_over_rsquared)
xlabel = plt.xlabel('Stellar Radius (R$_{\odot}$)') 
ylabel = plt.ylabel('dP/dr or -GM rho/r$^2$') 
```
